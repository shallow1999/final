```python
import shutil
import os
import cv2

# os.chdir(os.path.split(os.path.realpath(__file__))[0])
# print(os.getcwd())

dirs_path = "./video.mp4"  # 设置路径

# dirs = os.listdir(dirs_path)  # 获取指定路径下的文件
# for video_name in dirs:  # 循环读取路径下的文件并筛选输出
#     if os.path.splitext(video_name)[1] == ".mp4":  # 筛选mp4文件，只是筛选出第一个视频文件
#         print(video_name)  # 输出所有的mp4文件
#         break
video_name="./video.mp4"
output_img_path = './test_data/'
if not os.path.exists(output_img_path):
    os.mkdir(output_img_path)
else:
    # 删除原有目录
    shutil.rmtree(output_img_path)
    # 创建一个新目录
    os.mkdir(output_img_path)

cap = cv2.VideoCapture(video_name)
# videoFPS=cap.get(cv2.CAP_PROP_FPS)
# print (videoFPS)

cnt = 1

print("start")

while (1):
    ret, frame = cap.read()
    # show a frame
    if ret is True:
        print(frame.shape)
        cnt = cnt + 1
        print(cnt)   # 帧的数目
        if cnt % 1 == 0:   # 间隔两帧截取一张图片
            # cropped = frame[120:750, 260:1660]  # 裁剪坐标为[y0:y1, x0:x1]
            cropped = frame   # 没有经过裁剪，使用了原始的帧的图像。
            output_name = "%s%s%s" % (output_img_path, str(int(cnt / 2)), '.jpg')
            cv2.imwrite(output_name, cropped)
    else:
        break
print("finished")
```
